**Question 2**

$$
    f_{1}(x_1, x_2, x_3) = 2 (x_1 +x_2 + x_3 -3)^2 + (x_1-x_2)^2 + (x_2 - x_3)^2
$$
La fonction $f_{1}(x_1, x_2, x_3)$ est une fonction quadratique, ece qui signifie qu'elle peut approxim√©e par un polyn√¥me du deuxi√®me degr√©.

Calculons le gradient $\nabla f_{1}$ et la hessienne $\nabla^2 f_{1}$ de $f_{1}(x_1, x_2, x_3)$ :

$$
\nabla f_{1} = \begin{bmatrix}
4 (x_1 + x_2 + x_3 - 3) + 2 (x_1 - x_2) \\
4 (x_1 + x_2 + x_3 - 3) - 2 (x_1 - x_2) + 2 (x_2 - x_3) \\
4 (x_1 + x_2 + x_3 - 3) - 2 (x_2 - x_3)
\end{bmatrix}
$$

$$
\nabla^2 f_{1} = \begin{bmatrix}
6&2&4 \\
2&8&2 \\
4&2&6
\end{bmatrix}
$$

```julia
using Plots
using LinearAlgebra

# Fonction f1
f1(x1, x2, x3) = 2 * (x1 + x2 + x3 - 3)^2 + (x1 - x2)^2 + (x2 - x3)^2

# Gradient de f1
‚àáf1(x1, x2, x3) = [
    4 * (x1 + x2 + x3 - 3) + 2 * (x1 - x2),
    4 * (x1 + x2 + x3 - 3) - 2 * (x1 - x2) + 2 * (x2 - x3),
    4 * (x1 + x2 + x3 - 3) - 2 * (x2 - x3)
]

# Hessienne de f1
‚àá¬≤f1(x1, x2, x3) = [
    6  2  4;
    2  8  2;
    4  2  6
]

# Fixer certaines valeurs pour les autres variables
x2_fixed = 0
x3_fixed = 0

# D√©finir la plage de valeurs pour x1
x1_range = -5:0.1:5

# G√©n√©rer les donn√©es pour les trois graphes
f1_values = [f1(x1, x2_fixed, x3_fixed) for x1 in x1_range]
‚àáf1_values = [norm(‚àáf1(x1, x2_fixed, x3_fixed)) for x1 in x1_range]
‚àá¬≤f1_values = [tr(‚àá¬≤f1(x1, x2_fixed, x3_fixed)) for x1 in x1_range]

# La solution
x_sol = [1; 1; 1]

# G√©n√©rer les donn√©es pour les points
f1_x_sol = f1(x_sol[1], x_sol[2], x_sol[3])

# Cr√©er les graphes de courbes
plot(x1_range, f1_values, xlabel = "x‚ÇÅ", ylabel = "f‚ÇÅ", label = "f‚ÇÅ(x)")
plot!(x1_range, ‚àáf1_values, xlabel = "x‚ÇÅ", ylabel = "|‚àáf‚ÇÅ|", label = "‚àáf‚ÇÅ(x)")
plot!(x1_range, ‚àá¬≤f1_values, xlabel = "x‚ÇÅ", ylabel = "Tr(‚àá¬≤f‚ÇÅ)", label = "‚àá¬≤f‚ÇÅ(x)")

# Afficher les courbes dans le m√™me graphe 
plot!([x_sol[1]], [f1_x_sol], seriestype = :scatter, label = "Point solution", color = :red)
```
